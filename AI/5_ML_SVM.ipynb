```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools
import joblib
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error
from sklearn.metrics import recall_score, precision_score, f1_score, roc_auc_score, log_loss
from sklearn.metrics import roc_curve, auc
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
X = df.drop(['timestamp', 'datapath_id','flow_id', 'label'], axis=1)
y = df['label']
```

```python
X['ip_src'] = X['ip_src'].str.replace('.', '').astype('int64')
X['ip_dst'] = X['ip_dst'].str.replace('.', '').astype('int64')
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=28)

X_train = X_train.to_numpy()
X_val = X_val.to_numpy()
y_train = y_train.to_numpy()
y_val = y_val.to_numpy()
```

```python
print(f"Train: {X_train.shape} {y_train.shape}")
print(f"Val: {X_val.shape} {y_val.shape}")
```

```python
import time
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# === Scale the data nếu cần ===
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# X_train_scaled = scaler.fit_transform(X_train)
# X_val_scaled = scaler.transform(X_val)

# === Bắt đầu đếm thời gian huấn luyện ===
start_time = time.time()

# === Train mô hình SVM ===
svm_model = SVC(kernel='linear', probability=True, random_state=28, max_iter=1000)
svm_model.fit(X_train, y_train)

# === Dừng thời gian huấn luyện ===
end_time = time.time()

# === In ra thời gian huấn luyện ===
print(f"⏱️ Thời gian huấn luyện SVM: {end_time - start_time:.4f} giây")

# === Dự đoán (ví dụ) ===
# y_pred = svm_model.predict(X_val)
# print("Dự đoán xong.")
```

```python
# # Make predictions
start_pred = time.time()
y_pred = svm_model.predict(X_val)
y_proba = svm_model.predict_proba(X_val)[:, 1] # Get probability for the positive class (DDOS)
end_pred = time.time()
print(f"⏱️ Thời gian dự đoán: {end_pred - start_pred:.4f} giây")
```

```python
accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)
roc_auc = roc_auc_score(y_val, y_proba)
conf_matrix = confusion_matrix(y_val, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC: {roc_auc:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
cm = np.array([[61803, 210037],
               [29, 528388]])
# Optional: Visualize the confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

plt.figure()
plot_confusion_matrix(cm, classes=['Normal', 'DDOS'], title='Confusion matrix')
plt.show()
```
