```python
import csv
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import jaccard_score
```

```python
df = pd.read_csv('/content/dataset.csv')
```

```python
X = df.drop(['timestamp', 'datapath_id','flow_id', 'label'], axis=1)
y = df['label']
```

```python
X['ip_src'] = X['ip_src'].str.replace('.', '').astype('int64')
X['ip_dst'] = X['ip_dst'].str.replace('.', '').astype('int64')
```

```python
X.info()
```

```python
y.value_counts()
```

```python
print(f"X: {X.shape}")
print(f"y: {y.shape}")
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=28)

X_train = X_train.to_numpy()
X_val = X_val.to_numpy()
y_train = y_train.to_numpy()
y_val = y_val.to_numpy()
```

```python
print(f"Train: {X_train.shape} {y_train.shape}")
print(f"Val: {X_val.shape} {y_val.shape}")
```

```python
# prompt: Xây dựng mô hình dự đoán nhãn 0 với 1 của tập dữ liệu này, nghĩa là có DDOS hoặc không có DDOS, sử dụng mô hình NaiveBayes

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Khởi tạo mô hình Naive Bayes
model = GaussianNB()

# Huấn luyện mô hình trên tập huấn luyện
model.fit(X_train, y_train)


```

```python
# Dự đoán trên tập kiểm tra
y_pred = model.predict(X_val)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Tạo y_proba (giả sử bạn dùng RandomForestClassifier hoặc LogisticRegression)
y_pred = model.predict(X_val)
y_proba = model.predict_proba(X_val)[:, 1]  # Xác suất class 1

# Đánh giá mô hình
accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)
roc_auc = roc_auc_score(y_val, y_proba)
conf_matrix = confusion_matrix(y_val, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"ROC AUC: {roc_auc:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```
